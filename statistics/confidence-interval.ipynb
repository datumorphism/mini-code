```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(x, mu, sigma):
    return np.exp(-(x - mu)**2. / (2 * sigma**2 ))
```

```python
x=np.linspace(-3,3,10001)
mu_0 = 0
sigma_0 = 1
y=[ gaussian(i, mu_0, sigma_0) for i in x ]
```

```python
alpha_0 = 0.5
two_sigma = 2 * sigma_0
fig, ax = plt.subplots( figsize=(10,6.18) )
ax.plot(x,y, 'k-')
ax.set_ylim([0,1.1])
ax.set_yticks([])

ax.fill_between(x, 0, y, where= ( x < -two_sigma ), facecolor='red', alpha = alpha_0, interpolate=True)
ax.fill_between(x, 0, y, where= ( x > two_sigma ), facecolor='red', alpha = alpha_0, interpolate=True)
ax.fill_between(x, 0, y, where= ( [ (i >= -two_sigma) and (i<=two_sigma) for i in x] ), hatch="/", facecolor='green', alpha = alpha_0, interpolate=True)

ax.text(0, 0.5, r"$1-\alpha$", size=10, rotation=0.,
         ha="center", va="center",
         bbox=dict(boxstyle="round",
                   ec=(0.3, 0.3, 0.3),
                   fc=(0.7, 1, 0.7),
                   )
         )

ax.text(-2.2, 0.01, r"$\alpha/2$", size=10, rotation=0.,
         ha="center", va="bottom",
         bbox=dict(boxstyle="round",
                   ec=(1, 0.3, 0.3),
                   fc=(1., 0.6, 0.6),
                   )
         )
ax.text(2.2, 0.01, r"$\alpha/2$", size=10, rotation=0.,
         ha="center", va="bottom",
         bbox=dict(boxstyle="round",
                   ec=(1, 0.3, 0.3),
                   fc=(1., 0.6, 0.6),
                   )
         )
```
